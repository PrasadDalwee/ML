```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv("amazon2.csv")
df.head()
```

```python
def Search(query,search_type):
    index_arr=[]
    if search_type == 'description' or search_type == 'd':
        search='description'
    else:
        search='product_name'
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(df[search].fillna(''))
    query_vec = vectorizer.transform([query])
    results = cosine_similarity(X,query_vec).reshape((-1,))
    print(results)
    for i in results.argsort()[-10:][::-1]:
        print(i,'\n  ',df['product_name'].iloc[i],"\n  ",df['description'].iloc[i],'\n\n')
        index_arr.append(i)
    return index_arr
```

```python
query=input("Enter Query :")
search_type=input(" Enter Search Type  :")
# query='HORNBY Coach R4410A BR Hawksworth Corridor 3rd3 Hornby 00 Gauge BR Hawksworth 3rd Class W 2107 W # R4410A'
# search_type='d'
Search(query,search_type)
```
